```python
#CREATION OF LIST FOR BOARD 

def board_creator():
    global board
    board = []
    for i in range(0,9):
        board.append(" ")
    return board

################################################################################

#REPRESENTATION OF BOARD 

def display_board(board):
    
    clear_output    
    first_row = board[0] + " |" + board[1] + " |" + board[2]
    mid_one  = "__|__|__"
    mid_last = "  |  |  "
    second_row = board[3] + " |" + board[4] + " |" + board[5]
    third_row = board[6] + " |" + board[7] + " |" + board[8]
    
    print(first_row)
    print(mid_one)
    print(second_row)
    print(mid_one)
    print(third_row)
    print(mid_last)
        
################################################################################
   
    
#DEMO BOARD FOR HELP

def display_demo_board():
    
    
    mid_one  = "__|__|__"
    mid_last = "  |  |  " 
    
    print('1' + " |" + '2' + " |" + '3')
    print(mid_one)
    print('4' + " |" + '5' + " |" + '6')
    print(mid_one)
    print('7' + " |" + '8' + " |" + '9')
    print(mid_last)
    
################################################################################
    
#GAME INTRO FOR STARTING

def game_intro():
    print("Welcome to the game of Xs and Os. The rules are simple.")
    
    display_demo_board()
    
    
#################################################################################



#DECIDING OF SYMBOLS FOR PLAYERS
def choose_players():
        
        p1 = ""
        p2 = p1
        while p1 not in ['X','O']:
        
             p1 = input("You are player 1. Please choose your option. (X/O) ")
             print(f"You have chosen {p1}.")
             if p1 == "X":
                    p2 = "O"
             elif p1 =="O":
                    p2 = "X"
             else:
                    print("You have entered invalid input. ")
            
        return p1,p2
    

#################################################################################
#ENTERING OF CHOICE BY USER 


def choice_fn(p1):
    
    global allowed_range
    choice = " "
    print(f"It is {p1} turn")
    
    
    while choice not in allowed_range:
        choice = int(input("Please enter the position for marker. "))
        if (choice) not in allowed_range:
            print("You have entered invalid input.")
    allowed_range.remove(choice)
#FITTING OF CHOICE ON BOARD
    choice = choice - 1 
    board[choice] = p1
#RETURNING OF BOARD
    display_board(board)
    return board

###################################################################################

#CHECKING OF CONDITION 

def check_condition(board, player):
        global main_check 
        if ( (board[0] == player and board[1] == player and board[2] == player) or
        (board[0] == player and board[3] == player and board[6] == player) or
        (board[0] == player and board[4] == player and board[8] == player) or
        (board[1] == player and board[4] == player and board[7] == player) or 
        (board[2] == player and board[5] == player and board[8] == player) or 
        (board[2] == player and board[4] == player and board[6] == player) or
        (board[3] == player and board[4] == player and board[5] == player) or
        (board[6] == player and board[7] == player and board[8] == player) ) :  
            main_check = False
        return main_check

#####################################################################################
    
```

```python
def game_on():
    
    #INITIALIZATION
    global board
    global allowed_range
    global main_check
    main_check = True
    allowed_range = [1,2,3,4,5,6,7,8,9]
    board = board_creator()
    #board = board_creator()
    display_board(board)
    
    
     
    #GAME INTRO
    game_intro()
        
    #DECIDING OF SYMBOLS FOR PLAYERS
    p1, p2 = choose_players()
    
    #DISPLAYING OF PLAYERS
    print(f"Player 1 is {p1}. ")
    print(f"Player 2 is {p2}. ")
    
    count = 0
    #CHOICE TIME
    a = True
    while(a):

            
            #CHOICE TIME 1 
            updated_board = choice_fn(p1)
            a = check_condition(updated_board,p1)
            count += 1 
            print(count)
            if count >= 9:
                print("It is a draw. Thank you for playing.")
                a = False
                break
            if a == False:
                print(f"Congrats {p1}, you have won the game")
                break 


            #CHOICE TIME 2
            updated_board = choice_fn(p2)
            print(allowed_range)
                   
            
            a = check_condition(updated_board,p2)
            count += 1 
            if a == False:
                print(f"Congrats {p2}, you have won the game")
                break 
                
            
        

        
```
